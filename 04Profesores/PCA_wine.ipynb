```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_wine
```

```python
from scipy.stats import zscore

data = load_wine(as_frame=True)
df = data['data'].apply(zscore)
```

```python
df.head()
```

```python
pca = PCA()
pca.fit(df);
```

```python
plt.bar(range(13), pca.explained_variance_ratio_);
```

```python
sum(pca.explained_variance_ratio_[0:3])
```

```python
pca2 = PCA(n_components=2)
pca2.fit(df)
proj_data = pca2.transform(df)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(proj_data[:,0], proj_data[:,1], c='r', marker='.');
```
